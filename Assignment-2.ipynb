# 1.importing seaborn library

```python
import seaborn as sn
from matplotlib import pyplot as plt
```

### loading the car_crashes dataset which is present in seaborn datasets

```python
df=sn.load_dataset('car_crashes')
df.head()
```

### getting information about the dataset

```python
df.info()
```

```python
fig, axes = plt.subplots(2, 3,figsize=(18, 10))
sn.scatterplot(ax=axes[0, 0],x='total',y='speeding',data=df)
sn.scatterplot(ax=axes[0,1],x='total',y='alcohol',data=df)
sn.scatterplot(ax=axes[0,2],x='total',y='not_distracted',data=df)
sn.scatterplot(ax=axes[1,0],x='total',y='no_previous',data=df)
sn.scatterplot(ax=axes[1,1],x='total',y='ins_premium',data=df)
sn.scatterplot(ax=axes[1,2],x='total',y='ins_losses',data=df)
```

inference from above plots we can observe that by increasing speed,alcohol consumption,not_distracted rate and no previous accidents rates will increase the total rate of accidents.

```python
fig, axes = plt.subplots(2, 3,figsize=(18, 10))
sn.lineplot(ax=axes[0, 0],x="speeding",y="total",data=df)
sn.lineplot(ax=axes[0, 1],x="alcohol",y="total",data=df)
sn.lineplot(ax=axes[0, 2],x="not_distracted",y="total",data=df)
sn.lineplot(ax=axes[1, 0],x="no_previous",y="total",data=df)
sn.lineplot(ax=axes[1, 1],x="ins_premium",y="total",data=df)
sn.lineplot(ax=axes[1, 2],x="ins_losses",y="total",data=df)
```

from the above graphs we can see that line plots of total vs speed,alcohol,not-distracted,no-previous accidents.we can observe that at some points the slope is increasing and at some points it is decreasing.

```python
fig, axes = plt.subplots(2, 3,figsize=(18, 10))
sn.distplot(df["total"],ax=axes[0,0])
sn.distplot(df["speeding"],ax=axes[0,1])
sn.distplot(df["alcohol"],ax=axes[0,2])
sn.distplot(df["not_distracted"],ax=axes[1,0])
sn.distplot(df["no_previous"],ax=axes[1,1])
```

we can see that all the values are normally distributed because all the graphs are in bell shaped curve.

```python
corr=df.corr()
corr
```

from the above correlation table we can say that total and no previous accidents are strongly related.
speeding and alcohol are strongly related.
alcohol and total are strongly related.
not_distracted and total and no_previous and total are related.

```python
sn.heatmap(corr,annot=True,cmap="YlGnBu")
```
